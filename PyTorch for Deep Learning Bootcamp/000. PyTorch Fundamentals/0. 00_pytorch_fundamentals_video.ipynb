### Random Tensors

Why random tensors?

Random tensors are important because the way many neural networks learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data.

`Start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers`

Torch random tensor - https://pytorch.org/docs/stable/generated/torch.rand.html

```python
# Create a random tensor of size (3 ,4)

random_tensor = torch.rand(3, 4)

print(f"{random_tensor} \n")
print(f"{random_tensor.ndim} \n")
print(f"{random_tensor.shape}")
```

### Tensor datatypes
**Note:** Tensor datatypes is one of the 3 big errors you'll run into with PyTorch and deep learning:

1. Tensors not right datatype
2. Tensors not right shape
3. Tensors not on the right device

```python
# Float-32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=None,  # What datatype the tensor is (e.g. float32 or float16)
                               device=None,  # What device your tensor is on
                               requires_grad=False)  # Whether or not to track gradients with this tensor's operations

print(f"{float_32_tensor}")
print(f"{float_32_tensor.dtype}")
```

### Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning:

1. Element-wise multiplication
2. Matrix multiplication (dot product)

More information on multiplying matrices - https://www.mathsisfun.com/algebra/matrix-multiplying.html

There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match:
- `(3, 2) @ (3, 2)` won't work
- `(2, 3) @ (3, 2)` will work
- `(3, 2) @ (2, 3)` will work
2. The resulting matrix has the shape of the outer dimensions:
- `(2, 3) @ (3, 2)` -> `(2, 2)`
- `(3, 2) @ (2, 3)` -> `(3, 3)`

```python
torch.matmul(torch.rand(3, 2), torch.rand(2, 3))
```

```python
# Element-wise multiplication
print(tensor, "*", tensor)
print(f"Equals: {tensor * tensor}")
```

```python
%%time
# Matrix multiplication
torch.matmul(tensor, tensor)  # 1 + 4 + 9 = 14
```

```python
%%time
# Matrix multiplication by hand
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]

print(value)
```

## One of the most common errors in deep learning: shape errors

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 8],
                         [9, 10],
                         [11, 12]])

# torch.matmul(tensor_A, tensor_B)  ## This won't work because the inner dimensions don't match
```

To fix our tensor shape issues, we can manipulate the sahpe of one of our tensors using **transpose**.

A **transpose** switches the axes or dimensions of a given tensor.

```python
# The matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.T.shape}\n")  # Transposed using .T
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match\n")
print(f"Output: \n{torch.matmul(tensor_A, tensor_B.T)}\n")
print(f"Output Shape: {torch.matmul(tensor_A, tensor_B.T).shape}")
```

### Reshaping, stacking, squeezing, and unsqueezing tensors

- Reshaping - reshapes an input tensor to a defined shape
- View - returns a view of an input tensor of certain shape but keep the same memory as the original tensor
- Stacking - combine multiple tensor on top of each other (vertical stack) or side-by-side (horizontal stack)
- Squeeze - removes all `1` dimension from a tensor
- Unsqueeze - add a `1` dimension to a target tensor
- Permute  - return a view of the input with dimensions permuted (swapped) in a certain way

```python
# Let's create a tensor
x = torch.arange(1.0, 10.0)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the original input)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```python
# torch.squeeze() - removes all single dimensions from a target tensor
x_squeezed = x_reshaped.squeeze()
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}\n")

# Remove extra dimensions (removes all 1 dimensions) from x_reshaped
print(f"New tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
# torch.unsqueeze() - adds a single dimension to a target tensor at a specific dim (dimension)
print(f"Previous target: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}\n")

# Add an extra dimension with unsqueeze
print(f"Adding 1 dimension to the zeroth dimension: {x_squeezed.unsqueeze(dim=0)}")  # The zeroth dimension is the index 0/first axis, hence [1, 9]
print(f"New shape: {x_squeezed.unsqueeze(dim=0).shape}\n")
print(f"Adding 1 dimension to the first dimension: {x_squeezed.unsqueeze(dim=1)}")  # The first dimension is the index 1/second axis, hence [9, 1]
print(f"New shape: {x_squeezed.unsqueeze(dim=1).shape}")
```

```python
# torch.permute = rearranges the dimensions of a target tensor in a specificed order
x_original = torch.rand(size=(224, 224, 3))  # [height, width, color_channels]

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0 ,1)  # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")  # [color_channels, height, width]
```

```python
x_original[0, 0, 0] = 0.9999
x_original[0, 0, 0], x_permuted[0, 0, 0]
```

## PytTorch tensors and Numpy

NumPy is a popular scientific Python numerical computing library.

And becasue of this, PyTorch has functionality to interact with it.

- Converting data in NumPy to PyTorch tensor -> torch.from_numpy(nd_array)
- Converting PyTorch tensor to Numpy -> torch.Tensor.numpy()

```python
# NumPy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)  # NumPy has float64 as default type
tensor = torch.from_numpy(array)  # Torch.tensor has float32 as default type. However,
                                  # converting the numpy array to torch.tensor makes it also float64
array, array.dtype, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, tensor.dtype, numpy_tensor

# When converting between a NumPy array and a tensor, the resulting object will retain the original's data type
```

## Reproducibility (trying to take random out of random)

In short, how a neural network learns:
`Start with random numbers -> tensor operations -> update random numbers to try and make them better representations of data -> again -> again -> again...`

To reduce ithe randomness in neural networks and PyTorch comes the concep of a **random seed**.

Essentially what the random seed does is "flavor" the randomness.

```python
# Create two random tensors

random_tensor_A = torch.rand(3 ,4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some random but reproducible tensors

# Set the random_seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)

random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)  # Need set the seed again, otherwise, it is not the same seed anymore.
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
